```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('matplotlib-quattrocento.mplstyle')
```

```python
cet_1024 = np.loadtxt("CET1024.dat")
cet_512 = np.loadtxt("CET512.dat")
cet_256 = np.loadtxt("CET256.dat")
cet_128 = np.loadtxt("CET128.dat")
cet_64 = np.loadtxt("CET64.dat")

cet_64[:,1] = cet_64[:,1]*1e6
cet_128[:,1] = cet_128[:,1]*1e6
cet_256[:,1] = cet_256[:,1]*1e6
cet_512[:,1] = cet_512[:,1]*1e6
cet_1024[:,1] = cet_1024[:,1]*1e6

cet_64[:,2] = cet_64[:,2]*1e6
cet_128[:,2] = cet_128[:,2]*1e6
cet_256[:,2] = cet_256[:,2]*1e6
cet_512[:,2] = cet_512[:,2]*1e6
cet_1024[:,2] = cet_1024[:,2]*1e6
```

```python
fig = plt.figure(figsize=(6,4.5),dpi=150)
plt.plot(64*np.ones(len(cet_64)),cet_64[:,2],'s',label='64')
plt.plot(128*np.ones(len(cet_128)),cet_128[:,2],'s',label='128')
plt.plot(256*np.ones(len(cet_256)),cet_256[:,2],'s',label='256')
plt.plot(512*np.ones(len(cet_512)),cet_512[:,2],'s',label='512')
plt.plot(1024*np.ones(len(cet_1024)),cet_1024[:,2],'s',label='1024')

plt.xticks([64, 128, 256, 512, 1024]) #, 512, 1024])
plt.xlabel(r'Number of atoms')
plt.ylabel(r'Volumetric thermal expansion' '\n' r'coefficient at 300 K (10$^{-6}$K$^{-1}$)')
#plt.title(r'Filtered data')
#plt.legend(bbox_to_anchor=(0.9, 0.95), loc=1, borderaxespad=0.)
plt.show()
```

```python
data_to_plot = [cet_64[:,2], cet_128[:,2], cet_256[:,2], cet_512[:,2], cet_1024[:,2]]
fig = plt.figure(figsize=(6,4.5),dpi=150)
plt.boxplot(data_to_plot)
plt.xticks([1,2,3,4,5], ('64', '128', '256', '512', '1024'))
plt.xlabel(r'Number of atoms')
plt.ylabel(r'Volumetric thermal expansion' '\n' r'coefficient at 300 K (10$^{-6}$K$^{-1}$)')
plt.show()
fig.savefig('CET_300K_supercell_size.pdf',bbox_inches='tight')
```
