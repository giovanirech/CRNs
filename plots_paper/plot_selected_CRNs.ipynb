Selected CRNS chosen for dispersion, rendering and VvsT plot to be included in the paper

10% sp2 carbons: CRN 1175

70% sp2 carbons: CRN 0047

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def volume_fcn2(T, a0, a1, a2, a3, a4, a5, a6, a7, a8):
    x = T/1000
    return a0*np.exp( a1*a2/(np.exp(a2/x)-1)
                     +a3*a4/(np.exp(a4/x)-1)
                     +a5*a6/(np.exp(a6/x)-1)
                     +a7*a8/(np.exp(a8/x)-1))

params10 = [3217.65, 0.00589262, 0.506419, 0.0194865,1.50885,-0.00743811,3.36419,0,1]
params70 = [3699.64578, 0.00419844, 0.53270073,-0.00172756, 0.22845051, 0.01348754, 1.46877545, 0, 1]
```

```python
file10 = '../results/sp2/volume_vs_temperature/10_90/volume_vs_temperature_1175.fit'
M = np.loadtxt(file10)
T10 = M[:,0]
V10 = M[:,2]
V10_0 = 3217.63865 #taken from VT fitting, parameter a0

file70 = '../results/sp2/volume_vs_temperature/70_30/volume_vs_temperature_0047.fit'
M = np.loadtxt(file70)
T70 = M[:,0]
V70 = M[:,2]
V70_0 = 3699.60435 #taken from VT fitting, parameter a0
```

```python
T = np.linspace(0,300,50)
V10 = volume_fcn2(T, *params10)
V70 = volume_fcn2(T, *params70)
```

```python
plt.figure(dpi=150)
plt.rcParams['font.sans-serif']="Quattrocento Sans"
plt.rcParams['font.style']="normal"
plt.rcParams['font.size']=14

plt.plot(T,V10-V10[0],label='10% sp-C', color='k')
plt.plot(T,V70-V70[0],label='70% sp-C', color='C1')
plt.xlabel('Temperature (K)')
plt.ylabel('$\Delta$V ()')
plt.legend()
plt.savefig('volume_vs_temperature_selectedCRN.pdf', bbox_inches='tight')
plt.show()
```
