# Imports and definitions

```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

plt.rcParams['font.sans-serif']="Quattrocento Sans"
plt.rcParams['font.style']="normal"
plt.rcParams['font.size']=14

#Stoupin & Shvyd'ko, Phys. Rev. B 83, 104102
def empirical_stoupin(T):
    b = 3.6e-14
    c = 1.21e-11
    T0 = 212
    DT0 = 47
    
    W = 1.0/(1+np.exp((T-T0)/DT0))
    x = b*W*T**3.0 + c*(1-W)*T**2.0
    
    return 3.0*x
```

# Removing outliers before fitting

```python
#DATA WITHOUT OUTLIERS. We also generate a list with the ids of the CRNs to be included 


data_100 = [cet_10[:,0],cet_20[:,0],cet_30[:,0],cet_40[:,0],cet_50[:,0],cet_60[:,0],cet_70[:,0]]
data_300 = [cet_10[:,1],cet_20[:,1],cet_30[:,1],cet_40[:,1],cet_50[:,1],cet_60[:,1],cet_70[:,1]]
data_500 = [cet_10[:,2],cet_20[:,2],cet_30[:,2],cet_40[:,2],cet_50[:,2],cet_60[:,2],cet_70[:,2]]

ID = [id_10,id_20, id_30, id_40, id_50, id_60, id_70]

whis=1.5

selected = []
excluded = []
new_data_100 = []
new_data_300 = []
new_data_500 = []
for n in range(len(data_100)):
    cet100 = data_100[n]
    Q1_100,Q3_100 = np.percentile(cet100,[25,75])
    upper_limit_100 = Q3_100+whis*(Q3_100-Q1_100)
    lower_limit_100 = Q1_100-whis*(Q3_100-Q1_100)
    
    cet300 = data_300[n]
    Q1_300,Q3_300 = np.percentile(cet300,[25,75])
    upper_limit_300 = Q3_300+whis*(Q3_300-Q1_300)
    lower_limit_300 = Q1_300-whis*(Q3_300-Q1_300)
    
    cet500 = data_500[n]
    
    ids = []
    exc = []
    new_cet_100 = []
    new_cet_300 = []
    new_cet_500 = []
    for i in range(len(cet100)):
        if (lower_limit_100 < cet100[i] < upper_limit_100) and (lower_limit_300 < cet300[i] < upper_limit_300):
            ids.append(ID[n][i])
            new_cet_100.append(cet100[i])
            new_cet_300.append(cet300[i])
            new_cet_500.append(cet500[i])
        else:
            exc.append(ID[n][i])
    selected.append(ids)
    excluded.append(exc)
    new_data_100.append(new_cet_100)
    new_data_300.append(new_cet_300)
    new_data_500.append(new_cet_500)

print('Outliers removed for each sp2 ratio:', [len(x) for x in excluded])
```

# Mass density versus sp-C content

```python
v_diamond = np.loadtxt("../results/diamond/volume_vs_temperature.dat")[-1,1]

files_list = ["../results/sp2/volume_vs_temperature/volume_sp2_10_90.txt",
              "../results/sp2/volume_vs_temperature/volume_sp2_20_80.txt",
              "../results/sp2/volume_vs_temperature/volume_sp2_30_70.txt",
              "../results/sp2/volume_vs_temperature/volume_sp2_40_60.txt",
              "../results/sp2/volume_vs_temperature/volume_sp2_50_50.txt",
              "../results/sp2/volume_vs_temperature/volume_sp2_60_40.txt",
              "../results/sp2/volume_vs_temperature/volume_sp2_70_30.txt"]

volumes = [] #Each element in this list is a list with a sp2 ratio. 
for i, file in enumerate(files_list):
    lines = !cat  $file
    volumes.append(np.array([float(line.split()[-1]) for line in lines if line.split()[0] in ID[i]]))

#Check that everything looks fine
for i in range(len(ID)):
    print(len(ID[i]),len(volumes[i]))
```

```python
#Get specific mass from volumes

convertion_factor = 19.9452 #in g*/(cm*atom) for atoms of carbon 12.0107 a.m.u.

#diamond - 8 atoms in the unit cell
rho_0 = 8*convertion_factor/v_diamond
#CRNs - 512 atoms per unit cell
rho_10 = 512*convertion_factor/volumes[0]
rho_20 = 512*convertion_factor/volumes[1]
rho_30 = 512*convertion_factor/volumes[2]
rho_40 = 512*convertion_factor/volumes[3]
rho_50 = 512*convertion_factor/volumes[4]
rho_60 = 512*convertion_factor/volumes[5]
rho_70 = 512*convertion_factor/volumes[6]
```

```python
rho300_avg = [np.average(x) for x in [rho_10,rho_20,rho_30,rho_40,rho_50,rho_60,rho_70]]
rho300_std = [np.std(x,ddof=1)/np.sqrt(len(x)) for x in [rho_10,rho_20,rho_30,rho_40,rho_50,rho_60,rho_70]]

x = [10,20,30,40,50,60,70]

fig = plt.figure(dpi=150)
plt.errorbar(x, rho300_avg, yerr=rho300_std, label = '300 K', fmt='.:', capsize=3, capthick=1,elinewidth=1)
plt.scatter(0,rho_0)
plt.xlabel(r'sp content (%)')
plt.ylabel(r'Specific mass (g/cm)')
plt.show()
```

```python
from lmfit import Minimizer, fit_report, Parameters, Model
from lmfit.models import ParabolicModel

#DATA
x_exp = np.array([10,20,30,40,50,60,70])
y_exp = np.array(rho300_avg)
y_err = np.array(rho300_std)

modelo = LinearModel()
params = modelo.make_params()
result = modelo.fit(y_exp, params=params, x=x_exp,weights=1/y_err)


x_fit = np.linspace(x[0],x[-1])
y_fit = result.eval(result.params,x=x_fit)
y_err = result.eval_uncertainty(x=x_fit)

fig = plt.figure(dpi=150)
plt.plot(x_fit,y_fit,color='C1',zorder=1)
plt.fill_between(x_fit,y_fit+y_err, y_fit-y_err,edgecolor=None, facecolor='C1', alpha=0.1)
plt.errorbar(x, rho300_avg, yerr=rho300_std, fmt='o', capsize=3, color='k',
             capthick=1.5, elinewidth=1.5, markerfacecolor='w', markeredgecolor='k', zorder=2 )
#plt.scatter(0,rho_0,marker='s',color='C1', facecolor='w',linewidths=1.5) #diamond point
plt.xlabel(r'sp content (%)')
plt.ylabel(r'Specific mass (g/cm)')
plt.savefig('specificmass_vs_sp2_300K.pdf',bbox_inches='tight')
plt.show()

result
```
