```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

plt.style.use('/home/giovani/Dropbox/rhenium/rhenium_EOS/matplotlib-quattrocento.mplstyle')

n_atoms = 512
n_kpaths = 7
n_points_per_path = 20
labels_high_symmetry_points = ['½ 0 ½','0 0 ½','0 0 0','½ 0 0','½ ½ 0','½ ½ ½','0 ½ ½','0 ½ 0']
ticks_high_symmetry_points = np.arange(0,(n_points_per_path-1)*n_kpaths+1,(n_points_per_path-1))
```

```python
max_band = 3*n_atoms
ylimits = (0,500)

data = np.loadtxt('70_sp2_0047/70sp2.disp')
curves = np.reshape(data[:,1],(n_kpaths,n_points_per_path,n_atoms*3))
plt.figure(figsize=(5,8), dpi=150)
for i in range(0,max_band):
    for j in range(n_kpaths):
        x = (n_points_per_path-1)*j+np.arange(n_points_per_path)
        plt.plot(x,curves[j,:,i],'-k', lw=0.8, markersize=1)
        
plt.xlabel('K-path')
plt.ylabel('Frequency (cm$^{-1}$)')
plt.xlim(0,(n_points_per_path-1)*n_kpaths)
plt.vlines(ticks_high_symmetry_points,ylimits[0],ylimits[1], color='k', lw=0.5)
plt.ylim(ylimits)
plt.xticks(ticks=ticks_high_symmetry_points, labels=labels_high_symmetry_points, rotation=90)
plt.title('70% sp²  $-0.55\\times 10^{-6} K^{-1}$')
plt.tight_layout()
plt.savefig('70_sp2_dispersion.pdf')
plt.show()

#############################################################################

data = np.loadtxt('10_sp2_1175/10sp2.disp')
curves = np.reshape(data[:,1],(n_kpaths,n_points_per_path,n_atoms*3))
plt.figure(figsize=(5,8), dpi=150)
for i in range(0,max_band):
    for j in range(n_kpaths):
        x = (n_points_per_path-1)*j+np.arange(n_points_per_path)
        plt.plot(x,curves[j,:,i],'-k', lw=0.8, markersize=1)
        
plt.xlabel('K-path')
plt.ylabel('Frequency (cm$^{-1}$)')
plt.xlim(0,(n_points_per_path-1)*n_kpaths)
plt.vlines(ticks_high_symmetry_points,ylimits[0],ylimits[1], color='k', lw=0.5)
plt.ylim(ylimits)
plt.xticks(ticks=ticks_high_symmetry_points, labels=labels_high_symmetry_points, rotation=90)
plt.title('10% sp² $1.0\\times 10^{-6} K^{-1}$')
plt.tight_layout()
plt.savefig('10_sp2_dispersion.pdf')
plt.show()
```

```python
max_band = 3*n_atoms
ylimits = (0,500)

fig = plt.figure(figsize=(6,8), dpi=150)
grid = gridspec.GridSpec(ncols=4, nrows=1, figure=fig)
ax1 = fig.add_subplot(grid[0,:3])
ax2 = fig.add_subplot(grid[0,3])

# Dispersion curves
data = np.loadtxt('10_sp2_1175/10sp2_1175.disp')
curves = np.reshape(data[:,1],(n_kpaths,n_points_per_path,n_atoms*3))
for i in range(0,max_band):
    for j in range(n_kpaths):
        x = (n_points_per_path-1)*j+np.arange(n_points_per_path)
        ax1.plot(x,curves[j,:,i],'-k', lw=0.8, markersize=1)
        
ax1.set_xlabel('K-path')
ax1.set_ylabel('Frequency (cm$^{-1}$)')
ax1.set_xlim(0,(n_points_per_path-1)*n_kpaths)
ax1.vlines(ticks_high_symmetry_points,ylimits[0],ylimits[1], color='k', lw=0.5)
ax1.set_ylim(ylimits)
ax1.set_xticks(ticks=ticks_high_symmetry_points)
ax1.set_xticklabels(labels_high_symmetry_points, rotation=90)

# Density of states
freq_dos, dos= np.loadtxt('10_sp2_1175/10sp2_1175.dens', unpack=True)
ax2.fill_betweenx(y = freq_dos, x1 = dos, x2 = -0.001, lw=1,edgecolor=(0,0,0,1),facecolor=(0,0,0,0.2))
ax2.set_xlabel('DOS\n(1000/cm$^{-1}$)')
ax2.set_ylim(ylimits)
ax2.set_xlim(0.0,0.002)
ax2.set_yticks(ticks=[])
ax2.set_xticks(ticks=[0,0.001, 0.002])
ax2.set_xticklabels([0,1,2])

#plt.suptitle('10% sp² $1.0\\times 10^{-6} K^{-1}$')
plt.tight_layout(pad=0.4)
plt.savefig('10_sp2_1175_disp_dos.pdf')
plt.show()

##############################################################################

fig = plt.figure(figsize=(6,8),dpi=150)
grid = gridspec.GridSpec(ncols=4, nrows=1, figure=fig)
ax1 = fig.add_subplot(grid[0,:3])
ax2 = fig.add_subplot(grid[0,3])

# Dispersion curves
data = np.loadtxt('70_sp2_0047/70sp2_0047.disp')
curves = np.reshape(data[:,1],(n_kpaths,n_points_per_path,n_atoms*3))
for i in range(0,max_band):
    for j in range(n_kpaths):
        x = (n_points_per_path-1)*j+np.arange(n_points_per_path)
        ax1.plot(x,curves[j,:,i],'-k', lw=0.8, markersize=1)
        
ax1.set_xlabel('K-path')
ax1.set_ylabel('Frequency (cm$^{-1}$)')
ax1.set_xlim(0,(n_points_per_path-1)*n_kpaths)
ax1.vlines(ticks_high_symmetry_points,ylimits[0],ylimits[1], color='k', lw=0.5)
ax1.set_ylim(ylimits)
ax1.set_xticks(ticks=ticks_high_symmetry_points)
ax1.set_xticklabels(labels_high_symmetry_points, rotation=90)

# Density of states
freq_dos, dos= np.loadtxt('70_sp2_0047/70sp2_0047.dens', unpack=True)
ax2.fill_betweenx(y = freq_dos, x1 = dos, x2 = -0.001, lw=1,edgecolor=(0,0,0,1),facecolor=(0,0,0,0.2))
ax2.set_xlabel('DOS\n(1000/cm$^{-1}$)')
ax2.set_ylim(ylimits)
ax2.set_xlim(0.0,0.002)
ax2.set_yticks(ticks=[])
ax2.set_xticks(ticks=[0,0.001, 0.002])
ax2.set_xticklabels([0,1,2])

#plt.suptitle('70% sp² $-0.55\\times 10^{-6} K^{-1}$')
plt.tight_layout(pad=0.4)
plt.savefig('70_sp2_0047_disp_dos.pdf')
plt.show()
```
