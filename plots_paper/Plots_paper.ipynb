# Average CET as function of sp ratio

```python
a100_avg,a300_avg = np.average([cet_10, cet_20, cet_30, cet_40, cet_50, cet_60, cet_70],axis=1).transpose()
#stadard deviation of the mean
a100_std,a300_std = np.std([cet_10, cet_20, cet_30, cet_40, cet_50, cet_60, cet_70],axis=1,ddof=1).transpose()/np.sqrt(30)

#Insert diamond data
a100_avg = np.insert(a100_avg,0,cet_0[0])
a300_avg = np.insert(a300_avg,0,cet_0[1])
a100_std = np.insert(a100_std,0,0)
a300_std = np.insert(a300_std,0,0)

x = [0,10,20,30,40,50,60,70]

fig = plt.figure(dpi=150)
plt.errorbar(x, a100_avg, yerr=a100_std, label = '300 K', fmt='.:', capsize=3, capthick=1,elinewidth=1)
plt.xlabel(r'sp content (%)')
plt.ylabel(r'Volumetric thermal expansion' '\n' r'coefficient at 100 K (10$^{-6}$K$^{-1}$)')
plt.show()

fig = plt.figure(dpi=150)
plt.errorbar(x, a300_avg, yerr=a300_std, label = '300 K', fmt='.:', capsize=3, capthick=1,elinewidth=1)
plt.xlabel(r'sp content (%)')
plt.ylabel(r'Volumetric thermal expansion' '\n' r'coefficient at 300 K (10$^{-6}$K$^{-1}$)')
plt.show()
```

```python
from lmfit import Minimizer, fit_report, Parameters, Model
from lmfit.models import ParabolicModel
import numpy as np
import matplotlib.pyplot as plt


def exponential_function(x, a, b, c):
    return a*(x-b)**c

def parabola(x, a, b, c):
    print(x,a,b,c)
    return a*x**2.0+b*x+c

def func2min(parameters, x, y):
    return y-exponential_function(x,*[parameters[k].value for k in parameters.keys()])

#DATA
x_exp = np.array([0,20,30,40,50,60,70])
y_exp = np.array([3.812259 , 3.57794627, 3.7429737 , 3.9353108 , 4.3786772 , 5.04286003, 5.3845812 ])

parameters = Parameters()
# add with tuples:  (NAME VALUE VARY MIN  MAX  EXPR  BRUTE_STEP)
parameters.add_many(('a', 1.0, True, -100, 100, None, None),
                    ('b', -20, True,  -30.0, -10, None, None),
                    ('c', 1.0, True,  -1.0, 3.0, None, None))

minimizer = Minimizer(func2min,parameters, fcn_args=(x,a300_avg))
result = minimizer.minimize(method='ampgo')

x_fit = np.linspace(x[0],x[-1])
y_fit = exponential_function(x_fit,*[result.params[k].value for k in parameters.keys()])

fig = plt.figure(dpi=150)
plt.plot(x_fit,y_fit,color='C1')
plt.errorbar(x, a300_avg, yerr=a300_std, fmt='o', capsize=3, color='k',
             capthick=1.5, elinewidth=1.5, markerfacecolor='w', markeredgecolor='k' )

plt.xlabel(r'sp content (%)')
plt.ylabel(r'Volumetric thermal expansion' '\n' r'coefficient at 300 K (10$^{-6}$K$^{-1}$)')
plt.show()

result
```

```python
#DATA
y_err = np.array(a100_std)
y_err[0] = min(y_err[1:])/5.0

x_exp = np.array([0,10,20,30,40,50,60,70])


modelo = ParabolicModel()
params = modelo.make_params()
result = modelo.fit(a100_avg, params=params, x=x_exp,weights=1/y_err)


x_fit = np.linspace(x[0],x[-1])
y_fit = result.eval(result.params,x=x_fit)
y_err = result.eval_uncertainty(x=x_fit)

fig = plt.figure(dpi=150)
plt.plot(x_fit,y_fit,color='C1',zorder=1)
plt.fill_between(x_fit,y_fit+y_err, y_fit-y_err,edgecolor=None, facecolor='C1', alpha=0.1)
plt.errorbar(x, a100_avg, yerr=a100_std, fmt='o', capsize=3, color='k',
             capthick=1.5, elinewidth=1.5, markerfacecolor='w', markeredgecolor='k', zorder=2 )

plt.xlabel(r'sp content (%)')
plt.ylabel(r'Volumetric thermal expansion' '\n' r'coefficient at 100 K (10$^{-6}$K$^{-1}$)')
plt.savefig('CETvsSP2_100K_fit.pdf',bbox_inches='tight')
plt.show()

result
```

# Mass density versus sp-C content

```python
#cet_diamond = np.loadtxt('../VT_fittings/diamond/volume_vs_temperature.fit')
#cet_0  = cet_diamond[9,5]*1e6, cet_diamond[29,5]*1e6
v_0 = np.loadtxt("../diamond/volume_vs_temperature.dat")[-1,1]
v_10 = np.loadtxt("../VT_fittings/sp2/volume_sp2_10_90.txt")[1:31,3]
v_20 = np.loadtxt("../VT_fittings/sp2/volume_sp2_20_80.txt")[1:31,3]
v_30 = np.loadtxt("../VT_fittings/sp2/volume_sp2_30_70.txt")[1:31,3]
v_40 = np.loadtxt("../VT_fittings/sp2/volume_sp2_40_60.txt")[:31,3]
v_50 = np.loadtxt("../VT_fittings/sp2/volume_sp2_50_50.txt")[:31,3]
v_60 = np.loadtxt("../VT_fittings/sp2/volume_sp2_60_40.txt")[1:31,3]
v_70 = np.loadtxt("../VT_fittings/sp2/volume_sp2_70_30.txt")[1:31,3]

convertion_factor = 19.9452 #in g*/(cm*atom) for atoms of carbon 12.0107 a.m.u.

#diamond - 8 atoms in the unit cell
rho_0 = 8*convertion_factor/v_0
#CRNs - 512 atoms per unit cell
rho_10 = 512*convertion_factor/v_10
rho_20 = 512*convertion_factor/v_20
rho_30 = 512*convertion_factor/v_30
rho_40 = 512*convertion_factor/v_40
rho_50 = 512*convertion_factor/v_50
rho_60 = 512*convertion_factor/v_60
rho_70 = 512*convertion_factor/v_70
```

```python
rho300_avg = np.average([rho_10,rho_20,rho_30,rho_40,rho_50,rho_60,rho_70], axis=1)
rho300_std = np.std([rho_10,rho_20,rho_30,rho_40,rho_50,rho_60,rho_70], axis=1)

x = [10,20,30,40,50,60,70]

fig = plt.figure(dpi=150)
plt.errorbar(x, rho300_avg, yerr=rho300_std, label = '300 K', fmt='.:', capsize=3, capthick=1,elinewidth=1)
plt.scatter(0,rho_0)
plt.xlabel(r'sp content (%)')
plt.ylabel(r'Specific mass (g/cm)')
plt.show()
```

```python
from lmfit import Minimizer, fit_report, Parameters, Model
from lmfit.models import ParabolicModel

#DATA
x_exp = np.array([10,20,30,40,50,60,70])
y_exp = np.array(rho300_avg)
y_err = np.array(rho300_std)

modelo = ParabolicModel()
params = modelo.make_params()
result = modelo.fit(y_exp, params=params, x=x_exp,weights=1/y_err)


x_fit = np.linspace(x[0],x[-1])
y_fit = result.eval(result.params,x=x_fit)
y_err = result.eval_uncertainty(x=x_fit)

fig = plt.figure(dpi=150)
plt.plot(x_fit,y_fit,color='C1',zorder=1)
plt.fill_between(x_fit,y_fit+y_err, y_fit-y_err,edgecolor=None, facecolor='C1', alpha=0.1)
plt.errorbar(x, rho300_avg, yerr=rho300_std, fmt='o', capsize=3, color='k',
             capthick=1.5, elinewidth=1.5, markerfacecolor='w', markeredgecolor='k', zorder=2 )
#plt.scatter(0,rho_0,marker='s',color='C1', facecolor='w',linewidths=1.5) #diamond point
plt.xlabel(r'sp content (%)')
plt.ylabel(r'Specific mass (g/cm)')
plt.savefig('specificmass_vs_sp2_300K.pdf',bbox_inches='tight')
plt.show()

result
```
